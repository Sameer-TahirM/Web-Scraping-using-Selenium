Scrape 3 pages of reviews for a single product listing using product URL.<br /> Note: Product must have more than 1 review page. The code must be generic and should work for every product listing URL, with proper checks in place.

```python
#path for Google Chrome Driver in my System
path= "C:\Program Files (x86)\chromedriver.exe"
driver = webdriver.Chrome(path)

driver.get(url)
time.sleep(2)

# driver.maximize_window()
driver.execute_script("window.scrollTo(0, 1430)")

# main = driver.find_element_by_id('block-O28FlwAFD-')

for i in range(0,3):
    
    
    main = driver.find_element_by_id('block-O28FlwAFD-')
    
    x=1430
    # Function to scroll down a certain number of pixels
    driver.execute_script(f"window.scrollTo(0, {x})")
    time.sleep(1)
    
    #Finding the Next page Clickable Button
    try:
        next_page = main.find_element_by_xpath('/html/body/div[4]/div/div[9]/div[1]/div[2]/div/div/div/div[3]/div[2]/div/button[2]/i')
    
    except:
        pass
    

    people_review_block = main.find_element_by_class_name('mod-reviews')
    
    for i in people_review_block.find_elements_by_class_name('item'):
        
        try:
            # This is to check if there was an owner Review
            bot = i.find_element_by_class_name('item-title')


            n = i.find_element_by_class_name('middle')
            Name.append(n.text)

            r=i.find_element_by_class_name('item-content')
            b=i.find_element_by_class_name('content')
            Review.append(b.text)
            continue
            
            
        except:
            #Find the name of the user and his comment
            n = i.find_element_by_class_name('middle')
            Name.append(n.text)

            r=i.find_element_by_class_name('item-content')
            b=i.find_element_by_class_name('content')
            Review.append(b.text)
                   
    
    try:
        driver.execute_script("arguments[0].click();",next_page)
        time.sleep(2)
        
    except:
        driver.close()
    
    


driver.close()
```

```python
import pandas as pd
import numpy as np

# Removing 'by' and 'verified Purchaser'
for i in range(len(Name)):
    Name[i]=Name[i].replace('by', '')
    Name[i]=Name[i].replace('Verified Purchase', '')

#Creating a dataFrame of the Reviews!



Review_df = pd.DataFrame()

Review_df['Name'] = Name
Review_df['Review']=Review

# Review_df['Review'].replace('',np.nan,inplace=True)

Review_df.dropna(axis=0,inplace=True)

Review_df.reset_index(drop=True,inplace=True)

Review_df
```

### Question 1 Part b  (10 marks)

Clean the Reviews using NLTK,RE. Remove Punctuations, Smilies and other garbage using any library. 

Apply Vader Sentiment to find the sentiment of the review, and append it in a column named "Sentiment".

You can get help of Vader sentiment from this; https://towardsdatascience.com/sentimental-analysis-using-vader-a3415fef7664

```python
from nltk.corpus import stopwords
import string

# Getting stop words and punctutation

stopwords = set(stopwords.words('english'))
punc = list(string.punctuation)

final_r =[]

no_sp = []

user_r = Review_df['Review']

# Removing stop words and punctuation and check is the word is Alphabetical or a numeric (Removing garbage), also Lowering
for sent in user_r:
    tokens = sent.split()
    
    no_sp = [x.lower() for x in tokens if x not in stopwords and x not in punc and x.isalnum()]

    final_r.append( ' '.join(no_sp)   )

    
#creating a new df. That doesn't include garbage

Processed_df=pd.DataFrame()

Processed_df['Name']=Review_df['Name']
Processed_df['Review']=final_r

Processed_df
```

```python
# Finding Vader sentiment

from nltk.sentiment.vader import SentimentIntensityAnalyzer

sid = SentimentIntensityAnalyzer()

#Lambda Function to append each Row's sentiment
Processed_df['Sentiment'] = Processed_df['Review'].apply(lambda Review: sid.polarity_scores(Review))

Processed_df
```

Scrape 80 items for a keyword search e.g., 'mobile covers'<br/>Note: Must use "click()" function of webElement, URL changing for pages is not allowed. The code must be generic and should work for every keyword, with proper checks in place. Bonus for also scraping country and Daraz Mall availability.

```python
url = 'https://www.daraz.pk/'

path = "C:\Program Files (x86)\chromedriver.exe"
```

```python
query = input('Enter the Product you want to search for!\n')
Products_Requried=int(input('\nHow many products do u want?\n'))

Name =[]
Price=[]
Stars=[]
location=[]
Reviews =[]
DarPriv=[]


driver = webdriver.Chrome(path)

driver.get(url)

time.sleep(2)

driver.maximize_window()

#Finding the search bar
search = driver.find_element_by_id('q')

#Send the query into the search bar and pressing enter
search.send_keys(query)
search.send_keys(Keys.RETURN)


main_body = driver.find_element_by_class_name('box--ujueT')

no_ele = main_body.find_elements_by_class_name('gridItem--Yd0sa')

product_counter=0
scroll_by=100


#Done using Xpath
while product_counter <= Products_Requried:

    driver.execute_script("window.scrollTo(0, 250)")
    main_body = driver.find_element_by_class_name('box--ujueT')
    products_lst = main_body.find_elements_by_class_name('gridItem--Yd0sa')
    no_of_products = len(products_lst)
    
    for i in range(1,no_of_products):
        
        if(product_counter >= Products_Requried):
            break
            
        

        prod= driver.find_element_by_xpath(f'//*[@id="root"]/div/div[2]/div/div/div[1]/div[2]/div[{i}]')
        
        title=''
        try:# Try for Title
            title = prod.find_element_by_class_name('title--wFj93').text
            
        
        except: #except for Title
            title='None'
            pass
        
        Name.append(title)
            
        price=0    
        try:#Try for Price
            price = prod.find_element_by_class_name('price--NVB62').text
            price=price[4:]
            
        
        except:#Except for Price
            price=0
        
        Price.append(price)
        
        try:# try for reviews
            no_of_reviews = prod.find_element_by_class_name('rating__review--ygkUy').text
            Reviews.append(no_of_reviews[1:-1])
        
        except:# except for reviews
            Reviews.append(0)
        
        try:#Try for location
            loc = prod.find_element_by_class_name('location--eh0Ro').text
            if(loc=='Free Shipping'):
                loc='Overseas'

        except:#Except for location
            loc = 'No Location'

         #Append Location element   
        location.append(loc) 
        
        DarOrPriv=''
        try:#Try for Daraz Mall
            prod.find_element_by_xpath(f'//*[@id="root"]/div/div[2]/div/div/div[1]/div[2]/div[{i}]/div/div/div[2]/div[1]/i')
            DarOrPriv= 'Daraz Mall'
                
        except :#except for Daraz Mall
            DarOrPriv = 'Private'
            
        DarPriv.append(DarOrPriv)
        
        star=0.0
        
        try:#For stars
            prod.click()
            time.sleep(1)
            
            driver.execute_script("window.scrollBy(0,850)")
            time.sleep(1.7)
            
            
            driver.execute_script("window.scrollBy(0,(document.body.scrollHeight)/2.5)")
            time.sleep(2)
            star=0.0
            
            try:# Search for stars
                star = driver.find_element_by_class_name('score-average').text
                star = float(star)
                
            except:# Except Search for Star
                star=0.0
                
            
                
            
               
        except:#for Stars
            pass
        
        Stars.append(star)
        product_counter+=1   
        driver.back()
        time.sleep(1)
        
    if(product_counter >= Products_Requried):
        break    
        
    try:# Try for next page
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(1)

        xpath_nxt_page = driver.find_element_by_xpath('//*[@id="root"]/div/div[2]/div/div/div[1]/div[3]/div/ul/li[9]/a')
        driver.execute_script("arguments[0].click();",xpath_nxt_page)
        time.sleep(2)
        scroll_by=100
    
    except:#Except for next page
        break
            

driver.close()
```

```python
import pandas as pd

Daraz_df = pd.DataFrame()

Daraz_df['Name']=Name
Daraz_df['Price']=Price
Daraz_df['Stars']=Stars
Daraz_df['Reviews']=Reviews
Daraz_df['Location']=location
Daraz_df['Daraz Mall / Private']=DarPriv

#Removing ',' from prices

Daraz_df['Price'] = Daraz_df['Price'].str.replace(',','')

#Changing Price and Reviews into Integers

Daraz_df = Daraz_df.astype({'Price':'int32','Reviews':'int32'})


Daraz_df
```

```python
url = 'https://www.daraz.pk/'

path = "C:\Program Files (x86)\chromedriver.exe"
```

```python
query = input('Enter the kind of Product you want information!\n')

DarazOrPriv=[]
Avg_Price=[]
Avg_unit_selling=[]
Seller_Ratings=[]
Country=[]
Folowers=[]

lst_of_Sellers = []

def check_if_in_dict(name):
    if name in lst_of_Sellers:
        return True
    
    else:
        lst_of_Sellers.append(name)
        return False

driver = webdriver.Chrome(path)

driver.get(url)


time.sleep(2)

# driver.maximize_window()

#Finding the search bar
search = driver.find_element_by_id('q')

#Send the query into the search bar and pressing enter
search.send_keys(query)
search.send_keys(Keys.RETURN)

count=1
while(count<=10):
    
    for i in range(1,40):

            if(count>10):
                break

            prod= driver.find_element_by_xpath(f'//*[@id="root"]/div/div[2]/div/div/div[1]/div[2]/div[{i}]')
            prod.click()
            time.sleep(2.5)
            
            try:
                pop_up=driver.find_element_by_xpath('/html/body/div[6]/div/div[2]/div/span/i')
                pop_up.click()
                time.sleep(1)
                
            except:
                #No pop up continue
                time.sleep(1)
                pass

            name_of_seller = driver.find_element_by_xpath('//*[@id="module_seller_info"]/div/div[1]/div[1]/div[2]/a').text

            if check_if_in_dict(name_of_seller)==False:
                count+=1
                page=2
                
                #extract Price
                try:
                    price= driver.find_element_by_xpath('//*[@id="module_product_price_1"]/div/div/span').text
                    price= int(price[4:].replace(',',''))
                    
                except:
                    price=0

                Avg_Price.append(price)

                Avg_unit_selling.append(price/4-1)

                #Checking location based on shipping
                country=''
                try:
                    ship=driver.find_element_by_xpath('//*[@id="module_seller_delivery"]/div/div/div[2]/div[2]/div/div/div/div[1]/div').text
                    country='Overseas'

                except:
                    country='Pakistan'

                Country.append(country)


                #Click the profile
                view_prof=driver.find_element_by_xpath('//*[@id="module_seller_info"]/div/div[3]/a')
                view_prof.click()
                time.sleep(2)
                driver.execute_script("window.scrollTo(0, 50);")
                time.sleep(1)

                #Followers
                follow = driver.find_element_by_xpath('//*[@id="pi-component-container"]/div/div[2]/div/div/div/div[2]/div[1]/div/div/div/div[2]/div[2]/div').text
                follow=follow.split()
                try:
                    Folowers.append(int(follow[0]))
                except:
                    Folowers.append(0)

                # Daraz mall or Private
                try:
                    driver.find_element_by_xpath('//*[@id="pi-component-container"]/div/div[2]/div/div/div/div[2]/div[1]/div/div/div/div[2]/div[2]/img')
                    DorP='Daraz Mall'

                except:
                    DorP='Private'

                DarazOrPriv.append(DorP)

                #Rating of seller
                rating=driver.find_element_by_xpath('//*[@id="pi-component-container"]/div/div[2]/div/div/div/div[2]/div[1]/div/div/div/div[2]/div[3]').text
                rating=rating.split()
                rating=rating[0][:-1]
                try:
                    Seller_Ratings.append(int(rating))
                except:
                    Seller_Ratings.append(0)

                #Loop to go back to 1st page
                while(page!=0):
                    driver.back()
                    time.sleep(1)
                    page-=1

            else:# If name is already in the list. Go back
                driver.back()
                time.sleep(0.5)
                
    if(count>10):
        break
                
    try:# Try for next page
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(1)

        xpath_nxt_page = driver.find_element_by_xpath('//*[@id="root"]/div/div[2]/div/div/div[1]/div[3]/div/ul/li[9]/a')
        driver.execute_script("arguments[0].click();",xpath_nxt_page)
        time.sleep(2)
        scroll_by=100
    
    except:#Except for next page
        break
                    
            
time.sleep(1)                     
driver.close()
```
